```python
import pandas as pd
```

```python
users = pd.read_csv(r'C:\\Users\\sdp19\\Downloads\\u.user.txt', sep='|')     #Step3
```

```python
users.groupby('occupation').age.mean()  #Step4
```

```python
total_occupation_count = users.groupby(['occupation']).gender.count()
```

```python
filter_male_occupation_count = users.loc[users['gender'] == 'M']
```

```python
male_occupation_count = filter_male_occupation_count.groupby(['occupation', 'gender']).gender.count()
```

```python
result = (male_occupation_count/total_occupation_count)*100
```

```python
result.sort_values(ascending=False)  #Step5
```

```python
min_age, max_age = users.groupby(['occupation']).min('age'), users.groupby(['occupation']).max('age'); 
```

```python
min_age, max_age = min_age.rename(columns = {'age' : 'min_age'}), max_age.rename(columns = {'age' : 'max_age'}) #renaming the column name of age as min_age and max_age
```

```python
min_max_age = pd.concat([min_age, max_age], axis=1) 
```

```python
min_max_age.reset_index()[['occupation', 'min_age', 'max_age']] #Step6
```

```python
users.groupby(['occupation', 'gender']).age.mean()   #Step7
```

```python
occupation_count = users.groupby(['occupation', 'gender']).gender.count()
```

```python
result = round((occupation_count/total_occupation_count)*100, 2).rename('percentage')
```

```python
result = pd.DataFrame(result)
```
