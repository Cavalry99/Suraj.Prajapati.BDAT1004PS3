```python
import pandas as pd    #Step1
import numpy as np
import datetime as dt
```

```python
data = pd.read_csv(R'C:\\Users\\sdp19\\OneDrive\\Desktop\\wind.txt', delim_whitespace=True) #Step2
```

```python
data["Date"] = pd.to_datetime(data[["Yr","Mo","Dy"]].astype(str).agg('-'.join, axis=1))
data = data.drop(columns=["Yr","Mo","Dy"])        #Step3
data.head()
```

```python
data["Date"] = np.where(pd.DatetimeIndex(data["Date"]).year < 2000,data.Date,data.Date - pd.offsets.DateOffset(years=100)) #Step4
```

```python
newData = data.set_index("Date")
newData.index.astype("datetime64[ns]")   #Step5
```

```python
print(newData.isnull().values.ravel().sum())          #Step6
```

```python
x=newData.count()               
print("Total Non-missing values are :",x.sum())    #Step7
```

```python
y = newData.mean()
print('Mean windspeeds of the windspeeds over all the locations and all the times is', y.mean())     #Step8
```

```python
def stats(x):                          #Step9
    x = pd.Series(x)
    Min = x.min()
    Max = x.max()
    Mean = x.mean()
    Std = x.std()
    res = [Min,Max,Mean,Std]
    indx = ["Min","Max","Mean","Std"]
    res = pd.Series(res,index=indx)
    return res
loc_stats = newData.apply(stats)
loc_stats
```

```python
day_stats = newData.apply(stats,axis=1)    #Step10
day_stats.head()
```

```python
january_data = newData[newData.index.month == 1]
print ("January windspeeds:", january_data.mean())    #Step11
```

```python
print( "Yearly:", newData.resample('A').mean())       #Step12
```

```python
print ("Monthly:", newData.resample('M').mean())   #Step13
```

```python
print ("Weekly:", newData.resample('W').mean())   #Step14
```

```python
newdata = newData.groupby(lambda d: (d.month, d.year))
print ("Mean wind speed for each month in each location", newdata.mean()) #Step15
```
